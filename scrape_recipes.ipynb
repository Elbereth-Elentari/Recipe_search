```python
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import json
```

```python
current_links = []
with open("recipe_links.txt", "r") as current:
    for link in current:
        link = link.strip()
        current_links.append(link)
```

```python
#current_links = []
driver = webdriver.Chrome()
with open("recipe_links.txt", "a+") as recipe_links:
    for page_ending in range(69,91):
        driver.get("https://www.kwestiasmaku.com/home-przepisy?page=" + str(page_ending))
        hrefs_containers = driver.find_elements_by_class_name("field-content")
        for container in hrefs_containers:
            try:
                container = container.find_element_by_tag_name("a")
                href = container.get_attribute("href")
                if "/user/login" not in href and href not in current_links:
                    current_links.append(href)
                    recipe_links.write(href + "\n")
            except NoSuchElementException:
                pass
```

```python
recipes = []
driver = webdriver.Chrome()
for link in current_links[3524:]:
    driver.get(link)
    try:
        title = driver.find_element_by_css_selector("h1.przepis.page-header").text
        subtitle = driver.find_element_by_tag_name("h2").text
        name = title + ". " + subtitle
        categories_for_writing = []
        try:
            stars = driver.find_element_by_class_name("average-rating").text
            stars = stars.replace("Średnia: ", "")
            raters = driver.find_element_by_class_name("total-votes").find_element_by_tag_name("span").text
        except NoSuchElementException:
            stars = "0"
            raters = "0"
        try:
            likes = driver.find_element_by_class_name("count").text
            likes = likes.replace("(", "")
            likes = likes.replace(")", "")
        except NoSuchElementException:
            likes = "0"
        duration = ""
        difficulty = ""
        final_ingredients = []
        try:
            ingredient_container = driver.find_element_by_class_name("field-name-field-skladniki").find_elements_by_tag_name("li")
            for ingredient in ingredient_container:
                ingredient = ingredient.text
                final_ingredients.append(ingredient)
        except NoSuchElementException:
            pass
        final_instructions = []
        try:
            instructions = driver.find_element_by_class_name("field-name-field-przygotowanie").find_elements_by_tag_name("li")
            for instruction in instructions:
                instruction = instruction.text.strip()
                final_instructions.append(instruction + "\n")
        except NoSuchElementException:
            pass
        tags = []
        try:
            tags_container = driver.find_element_by_id("node-przepis-full-group-kategorie").find_elements_by_tag_name("a")
            for tag in tags_container:
                tag = tag.text.strip()
                tag = tag.lower()
                tags.append(tag)
        except NoSuchElementException:
            pass
        recipe = {"name" : name, "categories" : categories_for_writing, "rating" : stars, "raters" : raters, 
                       "likes": likes, "duration" : duration, "ingredients" : final_ingredients, "instructions" : 
                      final_instructions, "tags" : tags, "link" : link, "difficulty" : difficulty}
        recipes.append(recipe)
    except StaleElementReferenceException:
        pass
```

```python
with open("recipes_end.json", "w") as final:
    json.dump(recipes, final)
```
